```python
# for comparison between the numpy array and phython list
#using list
import time
list1=[i for i in range(10000000)]
list2=[i for i in range(10000000,20000000)]
list=[]
for i in range(len(list1)):
    start=time.time()
    list.append(list1[i]+list2[i])
print(list)
print(time.time()-start)
```

```python
#using numpy
import numpy as np
n1=np.arange(10000000)
n2=np.arange(10000000,20000000)
start=time.time()
n=n1+n2
print(time.time()-start)
# which numpy is fast due to direct store value and static structure
```

```python
# broadcasting concept for doing different shape arithmetic operations
e=np.arange(1,7).reshape(3,2)
r=np.arange(7,9).reshape(2)
print(e)
print(r)
print(e*r)
```

```python
# we can easily make advanced mathmatical function using numpy
def sigmoid(array):
    return 1/(1+np.exp(-array))
a=np.arange(10)
sigmoid(a)
```

```python
actual=np.arange(1,5)
predict=np.arange(6,10)
print(actual)
print(predict)
```

```python
# numpy is powerful for making custom function
def MSE(actual,predict):
    return np.mean((actual-predict)**2)
MSE(actual,predict)
```
